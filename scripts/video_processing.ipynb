```python
PATH_TO_VIDS = '/Users/adamcole/dev/of_v20220915_osx_release/apps/myApps/kissMeInteractive/bin/data/kiss_vids'
OUT_DIR = '/Users/adamcole/dev/of_v20220915_osx_release/apps/myApps/kissMeInteractive/bin/data/frameSequences2'
import os
from os import listdir
from os.path import isfile, join
import cv2

videos = [f for f in listdir(PATH_TO_VIDS) if isfile(join(PATH_TO_VIDS, f))]

for video in videos:
    video_path = f"{PATH_TO_VIDS}/{video}"
    seq_path, file_extension = os.path.splitext(f"{OUT_DIR}/{video}");
    if not os.path.exists(seq_path):
        os.makedirs(seq_path)
    
    vidcap = cv2.VideoCapture(video_path)
    success,image = vidcap.read()
    count = 0
    print(seq_path)
    while success:
        cv2.imwrite(f"${seq_path}/frame%d.jpg" % count, image)     # save frame as JPEG file      
        success,image = vidcap.read()
        print('Read a new frame: ', success)
        count += 1



```
